```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.plotly as py
from plotly.offline import download_plotlyjs,init_notebook_mode,plot, iplot
import plotly.graph_objs as go

init_notebook_mode(connected=True)

%matplotlib inline
```

```python
report2015 = pd.read_csv("world-happiness-report/2015.csv")
report2016 = pd.read_csv("world-happiness-report/2016.csv")
report2017 = pd.read_csv("world-happiness-report/2017.csv")
report2017.head()
```

```python
data = dict(
        type = 'choropleth',
        colorscale = 'Viridis',
        reversescale = True,
        locations = report2017['Country'],
        locationmode = "country names",
        z = report2017['Happiness.Rank'],
        text = report2017['Country'],
        colorbar = {'title' : 'Happiness Rank'},
      ) 

layout = dict(title = 'World Happiness Report 2017',
                geo = dict(showframe = False,projection = {'type':'Mercator'})
             )
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```
